```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
```

```python
# Q10: Produce a list of facilities with a total revenue less than 1000. 
# The output of facility name and total revenue, sorted by revenue.
# Remember that there's a different cost for guests and members!

def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT bf.name, SUM(bf.cost) AS revenue
FROM
(
SELECT * ,
CASE WHEN b.memid = 0 THEN f.guestcost * b.slots
ELSE f.membercost * b.slots
END
AS cost 
FROM Bookings AS b
INNER JOIN Facilities AS f
USING (facid)
) AS bf
GROUP BY bf.name
HAVING SUM(bf.cost) < 1000
ORDER BY revenue;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
# Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order

def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT sub.surname || ', ' || sub.firstname AS fullname, m1.surname || ', ' || m1.firstname AS recommendedby_fullname
FROM
(
SELECT * , CAST(m.recommendedby AS UNSIGNED) AS rec_int
FROM Members AS m
) AS sub
INNER JOIN Members AS m1
ON sub.rec_int = m1.memid
WHERE sub.rec_int NOT IN (0)
ORDER BY fullname;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
# Q12: Find the facilities with their usage by member, but not guests

def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """

SELECT sub.*, Members.surname || ', ' || Members.firstname AS fullname
FROM
(
SELECT f.name, b.memid, COUNT(*) AS total_usage
FROM Bookings AS b
INNER JOIN Facilities AS f
ON b.facid = f.facid
INNER JOIN Members AS m
ON b.memid = m.memid
WHERE m.firstname NOT IN ('GUEST')
GROUP BY b.memid, f.name
ORDER BY f.name, b.memid
) AS sub
LEFT JOIN Members
ON sub.memid = Members.memid;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
# Q13: Find the facilities usage by month, but not guests

def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
SELECT sub.name, sub.month, COUNT(*)
FROM
(
SELECT *,
CASE WHEN b.starttime LIKE '%-07-%' THEN '7'
WHEN b.starttime LIKE '%-08-%' THEN '8'
WHEN b.starttime LIKE '%-09-%' THEN '9'
END AS month
FROM Bookings AS b
INNER JOIN Facilities AS f
USING (facid)
) AS sub
INNER JOIN Members AS m
ON sub.memid = m.memid
WHERE m.firstname NOT IN ('GUEST')
GROUP BY sub.month, sub.name
ORDER BY sub.month, COUNT(*) DESC;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
